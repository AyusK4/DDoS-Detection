```python
import os
import pandas as pd

src_folder = r"..\CSV-01-12\01-12"  # Go up one directory, then into CSV-01-12/01-12
dst_folder = "."  # Save in current folder (10-1 attack-benign)
os.makedirs(dst_folder, exist_ok=True)

csv_files = [
    "DrDoS_DNS.csv",
    "DrDoS_LDAP.csv",
    "DrDoS_MSSQL.csv",
    "DrDoS_NetBIOS.csv",
    "DrDoS_NTP.csv",
    "DrDoS_SNMP.csv",
    "DrDoS_SSDP.csv",
    "DrDoS_UDP.csv",
    "Syn.csv",
    "TFTP.csv",
    "UDPLag.csv"
]

for file in csv_files:
    path = os.path.join(src_folder, file)
    print(f"Processing {file}...")
    benign_rows = []
    attack_rows = []
    # Read in chunks
    for chunk in pd.read_csv(path, chunksize=100000):
        benign_rows.append(chunk[chunk[' Label'] == 'BENIGN'])
        attack_rows.append(chunk[chunk[' Label'] != 'BENIGN'])
    # Concatenate all chunks
    benign_df = pd.concat(benign_rows)
    attack_df = pd.concat(attack_rows)
    # Keep all benign, sample 10x as many attack as benign (or all if not enough)
    n_benign = len(benign_df)
    n_attack = min(len(attack_df), n_benign * 10)
    attack_sampled = attack_df.sample(n=n_attack, random_state=42)
    # Combine and shuffle
    balanced_df = pd.concat([benign_df, attack_sampled]).sample(frac=1, random_state=42)
    # Save to current folder
    balanced_df.to_csv(os.path.join(dst_folder, file), index=False)
    print(f"Saved 10:1 {file} with {len(balanced_df)} rows ({n_benign} benign, {n_attack} attack).")
```

```python
import os
import pandas as pd

def merge_and_shuffle_csvs(folder_path, output_filename):
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    dfs = []
    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        try:
            df = pd.read_csv(file_path)
            dfs.append(df)
        except Exception as e:
            print(f"Could not read {file_path}: {e}")
    if dfs:
        merged_df = pd.concat(dfs, ignore_index=True).sample(frac=1, random_state=42).reset_index(drop=True)
        output_path = os.path.join(folder_path, output_filename)
        merged_df.to_csv(output_path, index=False)
        print(f"Saved merged file: {output_path}")
    else:
        print(f"No CSV files found in {folder_path}")

# Merge all 10-1 training samples into one CSV for training
#merge_and_shuffle_csvs("10-1 Training sample", "full-training-data.csv")
merge_and_shuffle_csvs("Test_80benign_20attack", "full-test-80-20.csv")
merge_and_shuffle_csvs("Test_60benign_40attack", "full-test-60-40.csv")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Load the merged training data
df = pd.read_csv("full-training-data.csv")

# Strip spaces from column names
df.columns = df.columns.str.strip().str.replace(' ', '')

# Drop identifier columns if present
id_cols = ['SourceIP', 'SourcePort', 'DestinationIP', 'DestinationPort', 'Timestamp', 'FlowID', 'Unnamed:0']
df.drop(columns=id_cols, inplace=True, errors='ignore')

# Drop columns with constant values
nunique = df.nunique()
constant_cols = nunique[nunique == 1].index.tolist()
df.drop(columns=constant_cols, inplace=True)

# Replace inf/-inf with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill NaN values: median for numeric, mode for object
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype in [np.float64, np.int64]:
            df[col].fillna(df[col].median(), inplace=True)
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)

# Convert Label to binary (1=attack, 0=benign)
if 'Label' in df.columns:
    df['Label'] = (df['Label'] != 'BENIGN').astype(int)

# Convert all columns to numeric if possible
for col in df.columns:
    if df[col].dtype == 'object':
        try:
            df[col] = pd.to_numeric(df[col])
        except:
            pass  # If not possible, consider encoding or dropping

# Encode any remaining object columns with LabelEncoder
encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le  # Save encoder for future use

# Show info
print(df.info())
print(df.head())
```

```python
from xgboost import XGBClassifier
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load preprocessed training data
df = pd.read_csv("Preprocessed/preprocessed_training.csv")
X = df.drop(columns=['Label'])
y = df['Label']

# Compute class weights (inverse of class frequency)
classes = np.unique(y)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y)
weight_dict = {cls: w for cls, w in zip(classes, class_weights)}
sample_weights = y.map(weight_dict)

# Train XGBoost on the full training set
model = XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    random_state=42,
    n_jobs=-1,
    verbosity=1,
    tree_method='hist'
)
model.fit(X, y, sample_weight=sample_weights)

# Feature importance plot
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]
feature_names = X.columns

plt.figure(figsize=(10, 6))
plt.title("XGBoost Feature Importances")
plt.bar(range(20), importances[indices][:20], align="center")
plt.xticks(range(20), feature_names[indices][:20], rotation=90)
plt.tight_layout()
plt.show()

# Print top 20 features
print("Top 20 important features:")
for i in range(20):
    print(f"{i+1}. {feature_names[indices[i]]}: {importances[indices[i]]:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report

# Load preprocessed training data
df = pd.read_csv("Preprocessed/preprocessed_training.csv")
X = df.drop(columns=['Label'])
y = df['Label']

# Get top 16 important features from previous XGBoost run
importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)
top_features = importance_df['feature'].head(16).tolist()

# Use only top features
X_top = X[top_features]

# Compute class weights
classes = np.unique(y)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y)
weight_dict = {cls: w for cls, w in zip(classes, class_weights)}
sample_weights = y.map(weight_dict)

# Train Random Forest
rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)
rf.fit(X_top, y)

# Train Decision Tree
dt = DecisionTreeClassifier(
    max_depth=10,
    random_state=42,
    class_weight='balanced'
)
dt.fit(X_top, y)

print("Random Forest and Decision Tree trained on top 16 features.")
```

```python
from sklearn.metrics import confusion_matrix

def false_positive_rate(y_true, y_pred):
    # Assuming '1' is attack, '0' is benign
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    return fp / (fp + tn)

# Random Forest FPR
rf_fpr_60_40 = false_positive_rate(y_test_60_40, rf.predict(X_test_60_40))
rf_fpr_80_20 = false_positive_rate(y_test_80_20, rf.predict(X_test_80_20))

# Decision Tree FPR
dt_fpr_60_40 = false_positive_rate(y_test_60_40, dt.predict(X_test_60_40))
dt_fpr_80_20 = false_positive_rate(y_test_80_20, dt.predict(X_test_80_20))

print(f"Random Forest FPR (60_40): {rf_fpr_60_40:.4f}")
print(f"Random Forest FPR (80_20): {rf_fpr_80_20:.4f}")
print(f"Decision Tree FPR (60_40): {dt_fpr_60_40:.4f}")
print(f"Decision Tree FPR (80_20): {dt_fpr_80_20:.4f}")
```

```python
# --- Extract statistics for top 12 features by attack type (with preprocessing) ---
import pandas as pd
import numpy as np
import os

# Top 12 features used in the model
top12_features = [
    'MinPacketLength', 'URGFlagCount', 'Inbound', 'Init_Win_bytes_forward',
    'min_seg_size_forward', 'ACKFlagCount', 'AveragePacketSize', 'FwdPacketLengthMean',
    'FwdPacketLengthMin', 'TotalBackwardPackets', 'TotalLengthofFwdPackets', 'FlowBytes/s'
]

# Attack CSV files in "10-1 Training sample"
attack_files = [
    "DrDoS_DNS.csv",
    "DrDoS_LDAP.csv",
    "DrDoS_MSSQL.csv",
    "DrDoS_NetBIOS.csv",
    "DrDoS_NTP.csv",
    "DrDoS_SNMP.csv",
    "DrDoS_SSDP.csv",
    "DrDoS_UDP.csv",
    "Syn.csv",
    "TFTP.csv",
    "UDPLag.csv"
]

# Folder containing raw attack files
folder = "10-1 Training sample"

# Dictionary to store statistics
attack_stats = {}

for file in attack_files:
    attack_name = file.replace(".csv", "")
    file_path = os.path.join(folder, file)
    
    print(f"Processing {attack_name}...")
    
    # Read CSV
    df_attack = pd.read_csv(file_path)
    
    # Strip spaces from column names
    df_attack.columns = df_attack.columns.str.strip().str.replace(' ', '')
    
    # Filter only attack traffic (not BENIGN)
    df_attack = df_attack[df_attack['Label'] != 'BENIGN']
    
    # --- Apply same preprocessing as in cell 3 ---
    # Drop identifier columns
    id_cols = ['SourceIP', 'SourcePort', 'DestinationIP', 'DestinationPort', 'Timestamp', 'FlowID', 'Unnamed:0']
    df_attack.drop(columns=id_cols, inplace=True, errors='ignore')
    
    # Drop constant columns
    nunique = df_attack.nunique()
    constant_cols = nunique[nunique == 1].index.tolist()
    df_attack.drop(columns=constant_cols, inplace=True)
    
    # Replace inf/-inf with NaN
    df_attack.replace([np.inf, -np.inf], np.nan, inplace=True)
    
    # Fill NaN values: median for numeric, mode for object
    for col in df_attack.columns:
        if df_attack[col].isnull().any():
            if df_attack[col].dtype in [np.float64, np.int64]:
                df_attack[col].fillna(df_attack[col].median(), inplace=True)
            else:
                df_attack[col].fillna(df_attack[col].mode()[0], inplace=True)
    
    # Convert all columns to numeric if possible
    for col in df_attack.columns:
        if df_attack[col].dtype == 'object' and col != 'Label':
            try:
                df_attack[col] = pd.to_numeric(df_attack[col])
            except:
                pass
    
    # Select only top 12 features that exist in this dataset
    available_features = [f for f in top12_features if f in df_attack.columns]
    df_features = df_attack[available_features]
    
    # Compute statistics
    stats = df_features.describe().T[['mean', '50%', 'min', 'max', 'std']]
    stats.columns = ['mean', 'median', 'min', 'max', 'std']
    stats['attack_type'] = attack_name
    
    attack_stats[attack_name] = stats
    print(f"  - Extracted stats for {len(available_features)} features from {len(df_features)} attack samples")

# Combine all statistics into one DataFrame
all_stats = pd.concat(attack_stats.values(), ignore_index=False)

# Save to CSV for easy reference
os.makedirs("attack_configs", exist_ok=True)
all_stats.to_csv("attack_configs/attack_statistics_top12.csv")

print("\n Attack statistics saved to: attack_configs/attack_statistics_top12.csv")
print("\nSample statistics (first 24 rows):")
print(all_stats.head(24))
```
