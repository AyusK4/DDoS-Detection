```python
# Balance the datasets by downsampling attack rows to match BENIGN rows

import os

src_folder = "CSV-01-12/01-12/"
dst_folder = "Balanced training set"
os.makedirs(dst_folder, exist_ok=True)

csv_files = [
    "DrDoS_DNS.csv",
    "DrDoS_LDAP.csv",
    "DrDoS_MSSQL.csv",
    "DrDoS_NetBIOS.csv",
    "DrDoS_NTP.csv",
    "DrDoS_SNMP.csv",
    "DrDoS_SSDP.csv",
    "DrDoS_UDP.csv",
    "Syn.csv",
    "TFTP.csv",
    "UDPLag.csv"
]

for file in csv_files:
    path = os.path.join(src_folder, file)
    print(f"Processing {file}...")
    benign_rows = []
    attack_rows = []
    # Read in chunks
    for chunk in pd.read_csv(path, chunksize=100000):
        benign_rows.append(chunk[chunk[' Label'] == 'BENIGN'])
        attack_rows.append(chunk[chunk[' Label'] != 'BENIGN'])
    # Concatenate all chunks
    benign_df = pd.concat(benign_rows)
    attack_df = pd.concat(attack_rows)
    # Sample attack rows to match benign count
    attack_sampled = attack_df.sample(n=len(benign_df), random_state=42)
    # Combine and shuffle
    balanced_df = pd.concat([benign_df, attack_sampled]).sample(frac=1, random_state=42)
    # Save to new folder
    balanced_df.to_csv(os.path.join(dst_folder, file), index=False)
    print(f"Saved balanced {file} with {len(balanced_df)} rows.")
```

```python


src_folder = "CSV-03-11/03-11/"
csv_files = [
    "LDAP.csv",
    "MSSQL.csv",
    "NetBIOS.csv",
    "Portmap.csv",
    "Syn.csv",
    "UDP.csv",
    "UDPLag.csv"
]

# Define your desired distributions as (benign%, attack%)
distributions = [
    (0.8, 0.2),  # 80% benign, 20% attack
    (0.6, 0.4)   # 60% benign, 40% attack
]

for benign_ratio, attack_ratio in distributions:
    folder_name = f"Test_{int(benign_ratio*100)}benign_{int(attack_ratio*100)}attack"
    dst_folder = os.path.join(src_folder, folder_name)
    os.makedirs(dst_folder, exist_ok=True)
    print(f"\nCreating test set in: {dst_folder}")

    for file in csv_files:
        path = os.path.join(src_folder, file)
        try:
            df = pd.read_csv(path)
            benign_df = df[df[' Label'] == 'BENIGN']
            attack_df = df[df[' Label'] != 'BENIGN']

            # Calculate number of samples for each class
            total_samples = min(len(benign_df) / benign_ratio, len(attack_df) / attack_ratio)
            n_benign = int(total_samples * benign_ratio)
            n_attack = int(total_samples * attack_ratio)

            # If not enough samples, use all available
            n_benign = min(n_benign, len(benign_df))
            n_attack = min(n_attack, len(attack_df))

            benign_sample = benign_df.sample(n=n_benign, random_state=42)
            attack_sample = attack_df.sample(n=n_attack, random_state=42)

            test_df = pd.concat([benign_sample, attack_sample]).sample(frac=1, random_state=42)
            test_df.to_csv(os.path.join(dst_folder, file), index=False)
            print(f"  {file}: {n_benign} benign, {n_attack} attack rows")
        except Exception as e:
            print(f"  {file} could not be processed: {e}")
```

```python


def merge_and_shuffle_csvs(folder_path, output_filename):
    csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    dfs = []
    for file in csv_files:
        file_path = os.path.join(folder_path, file)
        try:
            df = pd.read_csv(file_path)
            dfs.append(df)
        except Exception as e:
            print(f"Could not read {file_path}: {e}")
    if dfs:
        merged_df = pd.concat(dfs, ignore_index=True).sample(frac=1, random_state=42).reset_index(drop=True)
        output_path = os.path.join(folder_path, output_filename)
        merged_df.to_csv(output_path, index=False)
        print(f"Saved merged file: {output_path}")
    else:
        print(f"No CSV files found in {folder_path}")

# Merge for each folder
merge_and_shuffle_csvs("Balanced training set", "full-training-data.csv")
merge_and_shuffle_csvs("Test_80benign_20attack", "full-test-80-20.csv")
merge_and_shuffle_csvs("Test_60benign_40attack", "full-test-60-40.csv")
```

```python

import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns

# Load preprocessed training data
df = pd.read_csv('Preprocessed/preprocessed_training.csv')

# Split features and label
X = df.drop('Label', axis=1)
y = df['Label']

# Train XGBoost model
model = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
model.fit(X, y)

# Get feature importances
importances = model.feature_importances_
feature_names = X.columns
importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
importance_df = importance_df.sort_values(by='importance', ascending=False)

# Show top 20 features
top_n = 20
top_features = importance_df.head(top_n)

plt.figure(figsize=(10, 7))
sns.barplot(
    data=top_features,
    y='feature',
    x='importance',
    palette='viridis'
)
plt.title('Top 20 Feature Importances (XGBoost)', fontsize=16)
plt.xlabel('Importance', fontsize=14)
plt.ylabel('Feature', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Load preprocessed test set
df_test_60_40 = pd.read_csv('Preprocessed/preprocessed_test_60_40.csv')

# Split features and label
X_test = df_test_60_40.drop('Label', axis=1)
y_test = df_test_60_40['Label']

# Predict on test data
y_pred = model.predict(X_test)

# Accuracy
acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {acc:.4f}")

# Precision, Recall, F1
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
